```python
path = '/Volumes/ai_dev/input/ai_dataset/us-500.csv'

df = spark.read.format('csv').options(header='true').load(path)
display(df)

## header --> It uses the 1st line as name of column
## inferSchema --> It instruct spark to automatically infer the data types of each column based on the content of data
# df.printSchema() # used to display the schema of a DataFrame in a tree format
```

```python
df.printSchema()
```

#### 1) withColumn() --> function used to create new columns, convert the datatype of an existing column, update the value of existing column
#### 2) withColumnRenamed() --> function used to rename the column
#### 3) drop('col_name') --> function use to drop the column. drop('col1','col2','col3')--> drop multiple columns
#### 4) lit('constant_value') --> function is particularly useful when you want to add a new column with a constant value
#### 5) cast() --> used to convert the data type of a column to another datatype

```python
from pyspark.sql.functions import *

# create new columns
res = df.withColumn('Age',lit('25')).withColumn('Salary', lit('6000')).withColumn('Region',lit('Asia'))

# update the value of existing column
res = res.withColumn('Salary',col('Salary')+1000)

# convert data type of column to another data type
res = res.withColumn('zip',col('zip').cast('int')).withColumn('Salary',col('Salary').cast('int')).withColumn('Age',col('Age').cast('int'))

# Rename the column
res = res.withColumnRenamed('company_name','company').withColumnRenamed('email','email_address')

# Drop the columns
res = res.drop('phone1','Age')

display(res)
# res.printSchema()
```

```python
import re
path = '/Volumes/ai_dev/input/ai_dataset/10000Records.csv'
df = spark.read.format('csv').options(header='true',inferSchema='true').load(path)

# cleaning of columns
cols = [re.sub('[^a-zA-Z]','',column) for column in df.columns]
new_df = df.toDF(*cols)
# display(new_df)

# case when statement
res = new_df.withColumn('new_Gender',
                        when(col('Gender')=='F','Female')
                        .when(col('Gender')=='M','Male')
                        .otherwise('Invalid Gender')
                        ).withColumn('Bollywood_Name',
                                     when(col('FirstName').like('%li%'),'Salman').
                                     when(col('FirstName').like('%an%'),'Katrina').
                                     otherwise(col('FirstName')))

res = res.select('EmpID','Gender','new_Gender','FirstName','Bollywood_Name','Salary')

# count number of records
display(res.count())
# display(res)
```
